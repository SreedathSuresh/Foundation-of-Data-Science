```python
k
```

```python
k.isnull().sum()
```

```python
k.fillna(method='ffill',inplace=True)
k.isna().sum()
```

```python
k.shape
```

```python
cl = np.array(k['classification'])
cl
```

```python
for i in range(len(cl)):
    if cl[i]=='ckd\t':
        cl[i]='ckd'
```

```python
k['classification']=cl
k.classification.unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
k['classification']=le.fit_transform(k['classification'])
```

```python
k
```

```python
k.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
cols = k.columns
for i in cols:
    k[i]=ms.fit_transform(k[[i]])
```

```python
k.shape
```

```python
from sklearn.model_selection import train_test_split
y4 = k['classification']
k = k.drop('classification',axis=1)
x4 = k 
x_train,x_test,y_train,y_test = train_test_split(x4,y4,test_size=0.25,random_state=42)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_predk = knn.predict(x_test)
y_predk
```

```python
KnN = pd.DataFrame({"Predicted":y_predk,"Actual":y_test})
KnN
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

KnN_Confusion_matrix = confusion_matrix(y_test, y_predk)
sns.heatmap(KnN_Confusion_matrix, annot=True, fmt='g')
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_predk))
```

```python
#Svm
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
y_predS = model.predict(x_test)
y_predS
```

```python
Svm = pd.DataFrame({"Predicted":y_predS,"Actual":y_test})
Svm
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

Svm_Confusion_matrix = confusion_matrix(y_test, y_predS)
sns.heatmap(Svm_Confusion_matrix, annot=True, fmt='g')
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_predS))
```

```python
#Decision tree
from sklearn import tree
dc = tree.DecisionTreeClassifier()
dc.fit(x_train,y_train)
y_preddt = dc.predict(x_test)
y_preddt
```

```python
dt = pd.DataFrame({"Predicted":y_preddt,"Actual":y_test})
dt
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

Dt_Confusion_matrix = confusion_matrix(y_test, y_preddt)
sns.heatmap(Dt_Confusion_matrix, annot=True, fmt='g')
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_preddt))
```

```python
#Q7
h = pd.read_csv('/content/drive/MyDrive/Housing.csv')
h
```

```python
h.isnull().sum()
```

```python
ocols = h.describe(include='object').columns
for i in ocols:
    h[i]=le.fit_transform(h[[i]])
cols = h.columns

for i in cols[1:]:
    h[i]=ms.fit_transform(h[[i]])
```

```python
h.describe()
```

```python
hy = h['price']
h = h.drop('price',axis=1)
hx = h
hx_train,hx_test,hy_train,hy_test = train_test_split(hx,hy,test_size=0.2,random_state=42)
hx_train.shape, hx_test.shape, hy_train.shape, hy_test.shape
```

```python
from sklearn.linear_model import LinearRegression
Mr = LinearRegression()
Mr.fit(hx_train,hy_train)
Mr_pred = Mr.predict(hx_test)


from sklearn.neighbors import KNeighborsRegressor
KNR = KNeighborsRegressor(n_neighbors=7)
KNR.fit(hx_train,hy_train)
KNR_pred = KNR.predict(hx_test)
```

```python
plt.figure(figsize=(12,12))
plt.scatter(hy_test,Mr_pred)
plt.scatter(hy_test,KNR_pred)
plt.legend(['Multiple Linear Regression','KNN Regression'])
plt.xlabel('Actual value')
plt.ylabel('Predicted value')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
rmae = mean_absolute_error(hy_test,Mr_pred)
rmse = mean_squared_error(hy_test,Mr_pred)

kmae = mean_absolute_error(hy_test,KNR_pred)
kmse = mean_squared_error(hy_test,KNR_pred)

t = {'MAE':[rmae,kmae],'MSE':[rmse,kmse]}
t_df = pd.DataFrame(t)
t_df.index=['Multiple Linear Regression','KNN Regressor']
t_df
```

Mean Absolute Error is better than KNN and Mean Squared Error is better than Multiple Regression.
