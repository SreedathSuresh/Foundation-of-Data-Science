#Null Hypothesis:
All popular songs have high Danceability

#Alternative Hypothesis:
Popularity of a song is independent of Danceability

```python
d1 = pd.read_csv('/content/drive/MyDrive/spotify_top50_2021.csv')
d1.head()
```

```python
popmeanspl = d1['popularity'] > d1['popularity'].mean()
popmeanspl
```

```python
d1['popmeanspl'] = popmeanspl
d1
```

```python
pop_and_dance = d1[['popmeanspl', 'danceability']]
pop_and_dance
```

```python
gremean = pop_and_dance[ 'danceability'] [pop_and_dance['popmeanspl'] == True]
gremean
```

```python
lesmean = pop_and_dance[ 'danceability'] [pop_and_dance['popmeanspl'] == False]
lesmean
```

```python
pop_and_dance.hist(by = 'popmeanspl')
```

```python
import seaborn as sns
gremean.hist(histtype='stepfilled', alpha=.5, bins=30)  
lesmean.hist(histtype='stepfilled', alpha=.8, color=sns.desaturate("indianred", .75), bins=30)
plt.xlabel('Danceability',fontsize=15)
plt.ylabel('Song Popularity',fontsize=15)
plt.show()
```

It can be seen that many of the popular songs have lower danceability

```python
obs_diff = padmean['danceability'][1] - padmean['danceability'][0] 
obs_diff
```

```python
import array
diff = np.zeros(5000)
```

```python
for i in np.arange(5000):
    pop_and_dance = d1[['popmeanspl', 'danceability']]
    shuffled = pop_and_dance.sample(50,replace = False)
    shuffled_dnc = shuffled['danceability']
    original_and_shuffled = pop_and_dance.assign(shuffled_dnc=shuffled_dnc.values )
    group_means= original_and_shuffled.groupby('popmeanspl').mean()
    difference = group_means['shuffled_dnc'][0]- group_means['shuffled_dnc'][1]
    diff[i] = difference
diff_df = pd.DataFrame(diff)
diff_df
```

```python
diff_df.hist(bins = np.arange(-1,1,0.5))
plt.title('Prediction Under Null Hypotheses');
plt.xlabel('Differences between Group Averages',fontsize=15)
plt.ylabel('Units',fontsize=15);
print('Observed Difference:', obs_diff)
```

```python
np.count_nonzero(diff <= obs_diff)/diff.size
```

This shows that the popularity of a song is independent of the danceability of the song, ie, we reject the null hypothesis.
